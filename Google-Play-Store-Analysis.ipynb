## Content rating vs. size vs. rating – 3 variables at a time

```python
def size_to_bytes(size):
    if pd.isnull(size) or size == 'Varies with device':
        return np.nan
    size = size.strip().upper()
    if size.endswith('M'):
        return float(size[:-1]) * 1e6
    elif size.endswith('K'):
        return float(size[:-1]) * 1e3
    else:
        return float(size)

df['Size'] = df['Size'].apply(size_to_bytes)
```

### Create 5 buckets (20% records in each) based on Size

```python
#creating bins (buckets) of the given dataset and then we're giving labels to it. Basically creating the 5 gourps of the dataset

bins = [0, 20000000, 40000000, 60000000, 80000000, 100000000]  # Adjusted for bytes
labels = ['0-20MB', '20MB-40MB', '40MB-60MB', '60MB-80MB', '80MB-100MB']
df['Bucket'] = pd.cut(df['Size'], bins=bins, labels=labels)
pivot = pd.pivot_table(df, index='Bucket', values='Rating', columns='Content Rating')
print(pivot)
```

## By Content Rating vs. Size buckets, get the rating (20th percentile) for each combination

```python
quant=pd.pivot_table(df, values='Rating', index='Bucket', columns='Content Rating', aggfunc=lambda x:np.quantile(x,0.2))
quant
```

```python
# Make a heatmap of this

f,ax = plt.subplots(figsize=(7, 7))
sns.heatmap(quant, annot=True, linewidths=.5)
plt.show()
```
